```python
# Miguel Angel Martinez Moncada 1945469 Practica 5 Linear Models + correlation
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score

# Se lee el dataset preparado
df = pd.read_csv("Data Cleaning dataset-MAMM.csv")

# Se imprime el dataset
print(df)
```

```python
# Variables predictora y objetivo
X = df[['vendedor_1']]  # variable independiente
y = df['vendedor_2']    # variable dependiente

# Crear modelo
modelo = LinearRegression()
modelo.fit(X, y)

# Predicción
y_pred = modelo.predict(X)

# Valoe de R2
#R²	Interpretación
#0.00 No hay relación lineal
#0.30 Relación débil, pero detectable
#0.50 Relación moderada
#0.70 Buena capacidad explicativa
#1.00 Explicación perfecta (raro en datos reales)

r2 = r2_score(y, y_pred)
print(f"R² score del modelo: {r2:.4f}")
if r2 < 0.01:
    print("No hay relación lineal entre las variables.")
elif r2 < 0.3:
    print("La relación es muy débil. El modelo no explica bien la variabilidad.")
elif r2 < 0.5:
    print("La relación es débil. Puede ser útil con otras variables.")
elif r2 < 0.7:
    print("El modelo tiene una relación moderada. Aceptable para análisis preliminares.")
elif r2 < 0.9:
    print("Buena capacidad explicativa. El modelo es útil para predicción.")
else:
    print("Relación muy fuerte. El modelo explica muy bien los datos.")
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x='vendedor_1', y='vendedor_2', data=df, alpha=0.3, label='Datos reales')
plt.plot(df['vendedor_1'], y_pred, color='red', label='Línea de regresión')
plt.title('Regresión Lineal: Vendedor 1 vs Vendedor 2')
plt.xlabel('Ventas Vendedor 1')
plt.ylabel('Ventas Vendedor 2')
plt.legend()
plt.tight_layout()
plt.show()
```
